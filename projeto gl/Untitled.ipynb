```python
import pandas as pd
import re

# Tentar diferentes codificações
codificacoes = ['latin-1', 'iso-8859-1', 'cp1252', 'windows-1252']

for enc in codificacoes:
    try:
        with open("C:/Users/dell/Downloads/CLIENTES INATIVOS TXT.txt", 'r', encoding=enc) as arquivo:
            linhas = arquivo.readlines()
        print(f"✅ Conseguiu ler com encoding: {enc}")
        break
    except:
        print(f"❌ Falhou com: {enc}")
        continue

# Extrair dados
dados = []
for linha in linhas:
    linha = linha.strip()
    # Pular linhas vazias e cabeçalhos
    if not linha or 'Relatório' in linha or 'Somente entidades' in linha or 'Classificado' in linha:
        continue
    
    # Extrair código, nome, email e telefone
    match = re.match(r'^(\d+)\s+(.+?)\s+([\w\.\-]+@[\w\.\-]+)\s*(\(\d+\)[\d\-]+)?', linha)
    if match:
        codigo = match.group(1)
        nome = match.group(2).strip()
        email = match.group(3)
        telefone = match.group(4) if match.group(4) else ""
        dados.append([codigo, nome, email, telefone])

# Criar DataFrame e salvar
df = pd.DataFrame(dados, columns=['Codigo', 'Nome', 'Email', 'Telefone'])
df.to_csv("C:/Users/dell/Downloads/clientes_inativos.csv", index=False, encoding='utf-8-sig')

print(f"✅ Arquivo salvo com {len(df)} registros!")
df.head(50)
```
